```python
g = 9.8 # m/s^2
h0 = 10000 # 10 km
mu = 0.0035 # Person in free fall
#mu = 0.153 # Parachute
```

```python
t = 0
x = 0
v = 0
dt = 0.001
h = h0 - x
all_a = []
all_v = []
all_x = []
all_t = []
while t<40:
    a = g - mu*v**2
    t = t + dt
    v = v + a*dt
    x = x + v*dt
    h = h0 - x
    all_t.append(t)
    all_a.append(a)
    all_v.append(v)
    all_x.append(x)
```

```python
plot(all_t, all_v)
xlabel('Time [s]')
ylabel('Speed [m/s]')
grid()
```

```python
plot(all_t, all_x)
xlabel('Time [s]')
ylabel('Distance travelled [m]')
grid()
```

```python
plot(all_t, all_a)
xlabel('Time [s]')
ylabel('Acceleration [m/s$^2$]')
grid()
```

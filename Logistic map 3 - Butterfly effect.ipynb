#### This example shows :

- behaviour of logistic map for single value of supply parameter
- for two close initial conditions

In the chaotic regime the system diverges quickly even for  close initial parameters - this is the "Butterfly Effect"

```python
%pylab inline
```

```python
#r = 0.95 # Decay to zero
#r = 1.5 # Single value stable population
#r = 3.25 # Dual value stable population
#r = 3.55 # Quadruple value stable population
r = 3.95 # Chaos
x = asarray([ 0.5, 0.50000001 ])
all_x = []
for i in range(1000):
    all_x.append(x)
    x = r*x*(1-x)
```

```python
plot(all_x)
axis([-1,50,-0.02,1.02])
grid()
xlabel('Generation, $i$')
ylabel('Population, $x_i$')
```

```python
plot(all_x)
axis([950,1000,-0.02,1.02])
grid()
xlabel('Generation, $i$')
ylabel('Population, $x_i$')
```

```python
plot(column_stack(all_x)[0,:]- column_stack(all_x)[1,:])
grid()
```

```python
plot(column_stack(all_x)[0,:], column_stack(all_x)[1,:],'x')
grid()
xlabel('Population 1, $x^A_i$')
ylabel('Population 2, $x^B_i$')
axis('square');
```

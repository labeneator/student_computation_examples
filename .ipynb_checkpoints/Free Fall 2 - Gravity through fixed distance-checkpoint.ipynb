#### This example shows :

- free fall in constant gravity field with acceleration "g" close to Earth 
- from release at fixed height "h0" until ground level is reached
- computing speed and distance travelled as function of time

```python
%pylab inline
```

```python
g = 9.8 # m/s^2
h0 = 10000 # 10 km
```

```python
t = 0
x = 0
v = 0
dt = 0.001
h = h0 - x
all_a = []
all_v = []
all_x = []
all_t = []
while h>0:
    a = g
    t = t + dt
    v = v + a*dt
    x = x + v*dt
    h = h0 - x
    all_t.append(t)
    all_a.append(a)
    all_v.append(v)
    all_x.append(x)
```

```python
plot(all_t, all_v)
xlabel('Time [s]')
ylabel('Speed [m/s]')
grid()
```

```python
plot(all_t, all_x)
xlabel('Time [s]')
ylabel('Distance travelled [m]')
grid()
```

```python
plot(all_t, h0-asarray(all_x))
xlabel('Time [s]')
ylabel('Height [m]')
grid()
```

```python
plot(all_t, all_a)
xlabel('Time [s]')
ylabel('Acceleration [m/s$^2$]')
grid()
```

```python
g = 9.8 # m/s^2
h0 = 10000 # 10 km
mu0_person = 0.0035 # Person in free fall
mu0_parachute = 0.153 # Parachute
dt_open = 3 # s
Hs = 8500
```

```python
t = 0
x = 0
v = 0
dt = 0.001
h = h0 - x
all_a = []
all_v = []
all_x = []
all_t = []
all_mu = []
while h>0:
    if(h>1000):
        mu0 = mu0_person
        t_open = t
    elif t < (t_open+dt_open):
        mu0 = mu0_person*(t_open+dt_open-t)/dt_open + mu0_parachute*(t-t_open)/dt_open
    else:
        mu0 = mu0_parachute
    mu = mu0 * exp(-h/Hs)
    a = g - mu*v**2
    t = t + dt
    v = v + a*dt
    x = x + v*dt
    h = h0 - x
    all_t.append(t)
    all_a.append(a)
    all_v.append(v)
    all_x.append(x)
    all_mu.append(mu)
```

```python
plot(all_t, all_v)
xlabel('Time [s]')
ylabel('Speed [m/s]')
grid()
```

```python
plot(all_t, all_x)
xlabel('Time [s]')
ylabel('Distance travelled [m]')
grid()
```

```python
plot(all_t, h0-asarray(all_x))
xlabel('Time [s]')
ylabel('Height [m]')
grid()
```

```python
plot(all_t, all_a)
xlabel('Time [s]')
ylabel('Acceleration [m/s$^2$]')
grid()
```

```python
plot(all_t, all_mu)
xlabel('Time [s]')
ylabel('Resistance factor [1/m]')
grid()
```

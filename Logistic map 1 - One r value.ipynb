#### This example shows :

- behaviour of logistic map for single value of supply parameter

```python
%pylab inline
```

```python
#r = 0.95 # Decay to zero
#r = 1.5 # Single value stable population
#r = 3.25 # Dual value stable population
#r = 3.55 # Quadruple value stable population
r = 3.95 # Chaos
x = 0.5
all_x = []
for i in range(1000):
    all_x.append(x)
    x = r*x*(1-x)
```

```python
plot(all_x)
axis([-1,50,-0.02,1.02])
grid()
xlabel('Generation, $i$')
ylabel('Population, $x_i$')
```

```python
plot(all_x)
axis([950,1000,-0.02,1.02])
grid()
xlabel('Generation, $i$')
ylabel('Population, $x_i$')
```

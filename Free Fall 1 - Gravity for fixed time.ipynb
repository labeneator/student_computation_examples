```python
g = 9.8 # m/s^2
```

```python
t = 0
x = 0
v = 0
dt = 0.001
all_a = []
all_v = []
all_x = []
all_t = []
while t<10:
    a = g
    t = t + dt
    v = v + a*dt
    x = x + v*dt
    all_t.append(t)
    all_a.append(a)
    all_v.append(v)
    all_x.append(x)
```

```python
plot(all_t, all_v)
xlabel('Time [s]')
ylabel('Speed [m/s]')
grid()
```

```python
plot(all_t, all_x)
xlabel('Time [s]')
ylabel('Distance travelled [m]')
grid()
```

```python
plot(all_t, all_a)
xlabel('Time [s]')
ylabel('Acceleration [m/s$^2$]')
grid()
```

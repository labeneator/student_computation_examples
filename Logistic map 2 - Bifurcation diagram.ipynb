#### This example shows :

- behaviour of logistic map for a range of values of supply parameter
- generating the famous bifurcation diagram

```python
%pylab inline
```

```python
r = arange(0.0,4.0,0.001)
x = r*0 + 0.1
all_x = []
for i in range(1000):
    if(i>900):
        all_x.append(x)
    x = r*x*(1-x)
```

```python
figure(figsize=(10,6))
plot(r,column_stack(all_x),'k,')
axis([-0.1,4.1,-0.05,1.05])
grid()
xlabel('Supply parameter, r')
ylabel('Population, x')
```
